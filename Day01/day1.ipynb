```python
def part1():
  dial = 100
  pos = 50
  zeros = 0
  with open("inputtest.txt") as f:
    x = f.read().splitlines()
    for turn in x:
      tpos = int(turn[1:])
      if turn[0] == "L":
        pos = ((pos - tpos) + dial) % dial
      else:
        pos = (pos + tpos) % dial
      if pos == 0:
        zeros += 1

  print(zeros)
  
```

```python
def part2():
  dial = 100
  pos = 50
  zeros = 0
  with open("input.txt") as f:
    x = f.read().splitlines()
    for turn in x:
      tpos = int(turn[1:])
      for _ in range(tpos):  # Keep turning at some point it passes 0 or 100 which MOD 100 equals 0.
        if turn[0] == "L":
          # pos = ((pos - tpos) + dial) % dial
          pos = (pos - 1 + dial) % dial
        else:
          # pos = (pos + tpos) % dial
          pos = (pos + 1) % dial
        if pos == 0:
          zeros += 1

  print(zeros)
  
```
